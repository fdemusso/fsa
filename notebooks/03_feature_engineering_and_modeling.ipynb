## Phase 1: Metric Synthesis & Ground Truth Construction

To operationalize the qualitative component of our hybrid engine, we must first transition from **review-level granularity** to **movie-level metrics**. This block performs a statistical reduction of the raw critic data, collapsing thousands of individual sentiment scores into a stable central tendency, $S_{final\_mean}$. This establishes the primary qualitative anchor required for our subsequent normalization and ranking synthesis.

```python
import pandas as pd
reviews_df['final_score'] = pd.to_numeric(reviews_df['final_score'], errors='coerce')
avg_scores = reviews_df.groupby('rotten_tomatoes_link')['final_score'].mean()
movies_df['s_final_mean'] = movies_df['rotten_tomatoes_link'].map(avg_scores)
```
